```python
from sklearn.ensemble import RandomForestClassifier

y = titanic["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(titanic[features])
X_test = pd.get_dummies(titanic1[features])

model = RandomForestClassifier(n_estimators=500, max_depth=5, random_state=0) ## Modelo de entrenamiento cambiado en el numero de estimadores

model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': titanic1.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
